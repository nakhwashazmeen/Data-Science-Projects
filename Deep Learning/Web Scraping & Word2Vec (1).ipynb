```python
#Data Handling Packages
import pandas as pd
import numpy as np

#Web Scraping
import bs4 as bs
import urllib.request
import requests

# Text Cleaning libraries
import re
import nltk
from nltk.corpus import stopwords
stops = stopwords.words("english")

# Word2Vec
from gensim.models import Word2Vec
```

```python
nlp_url = "https://en.wikipedia.org/wiki/Natural_language_processing"
data_nlp = urllib.request.urlopen(nlp_url)
```

```python
data_nlp = data_nlp.read()
```

```python
data_nlp = bs.BeautifulSoup(data_nlp, "xml")
```

```python
data_nlp
```

```python
paragraph_nlp = data_nlp.find_all("p")
```

```python
article_nlp = " "
for p in paragraph_nlp:
    article_nlp = article_nlp + p.text
```

```python
article_nlp
```

```python
data_machine_learning_url = "https://en.wikipedia.org/wiki/Machine_learning"
data_machine_learning = urllib.request.urlopen(data_machine_learning_url)
```

```python
data_machine_learning = data_machine_learning.read()
```

```python
data_machine_learning
```

```python
data_machine_learning = bs.BeautifulSoup(data_machine_learning, "xml")
```

```python
paragraph_ML = data_machine_learning.find_all("p")
```

```python
article_ML = " "
for p in paragraph_ML:
    article_ML = article_ML + p.text
```

```python
article_ML
```

```python
# Merging all the articles
articles = article_deep_learning + article_nlp + article_ML
```

```python
len(articles)
```

```python
#Text Cleaning
def clean_text(text):
    text = text.lower()
    text = re.sub("[^a-z]"," ",text)
    word = text.split()
    imp_word = [w for w in word if w not in stops]
    return [imp_word]
```

```python
clean_article = clean_text(articles)
```

```python
clean_article
```

```python
len(clean_article[0])
```

## Word2Vec

```python
window_length = 5
embedding_dimensions = 300
epoch = 10
```

```python
word2vec = Word2Vec(sentences=clean_article, vector_size=embedding_dimensions, window=window_length, epochs=epoch, sg=1)
```

```python
word = word2vec.wv.index_to_key
```
